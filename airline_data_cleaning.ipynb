```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/MLBA/BA476 Project/old notebooks/train.csv',index_col=0)
test_df = pd.read_csv('/content/drive/MyDrive/MLBA/BA476 Project/old notebooks/test.csv',index_col=0)
full_df = pd.concat([train_df,test_df])
full_df
```

```python
full_df.info()
```

```python
clean_df = full_df.dropna(how='any')
numerical_cols = clean_df.select_dtypes(include=['float','int']).columns
categorical_cols = clean_df.columns.difference(numerical_cols).drop('satisfaction')
categorical_cols
```

```python
dummies = pd.get_dummies(clean_df[categorical_cols],prefix=categorical_cols)
clean_df = pd.concat([clean_df,dummies],axis=1)
clean_df
```

```python
clean_df['satisfaction'] = clean_df['satisfaction'].apply(lambda x: x == 'satisfied')
clean_df
```

```python
final_df = clean_df.drop(categorical_cols,axis = 1)
final_df.columns
```

```python
final_df
```

```python
final_df.reset_index(drop=True).to_csv('/content/drive/MyDrive/MLBA/BA476 Project/airline_data_clean_final.csv')
```
